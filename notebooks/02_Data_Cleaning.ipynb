## 2.1 Setup

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load raw data
df = pd.read_csv('/content/drive/MyDrive/CarPrice_Assignment.csv')
print(f"Loaded: {df.shape}")
df.head()
```

## 2.2 Extract Brand Name from CarName

The `CarName` column contains both brand and model (e.g., "toyota corolla"). We extract the brand and fix common typos.

```python
# Extract brand (first word of CarName)
df['brand'] = df['CarName'].str.split(' ').str[0].str.lower().str.strip()

# Check unique brands before cleaning
print(f"Unique brands before cleaning: {df['brand'].nunique()}")
print(sorted(df['brand'].unique()))
```

```python
# Typo correction dictionary
brand_fix = {
    'alfa-romero': 'alfa-romeo',
    'maxda': 'mazda',
    'nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}

df['brand'] = df['brand'].replace(brand_fix)

print(f"Unique brands after cleaning: {df['brand'].nunique()}")
print(sorted(df['brand'].unique()))
```

```python
# Brand value counts
df['brand'].value_counts()
```

## 2.3 Drop Unnecessary Columns

```python
# Drop car_ID (arbitrary index) and CarName (replaced by brand)
df.drop(columns=['car_ID', 'CarName'], inplace=True)
print(f"Shape after dropping car_ID and CarName: {df.shape}")
```

## 2.4 Check and Fix Data Types

```python
df.dtypes
```

```python
# Convert cylindernumber from word to numeric
cylinder_map = {
    'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'eight': 8, 'twelve': 12
}
df['cylindernumber'] = df['cylindernumber'].map(cylinder_map)
print("Cylinder number converted to numeric:")
print(df['cylindernumber'].value_counts().sort_index())
```

```python
# Convert doornumber from word to numeric
door_map = {'two': 2, 'four': 4}
df['doornumber'] = df['doornumber'].map(door_map)
print("Door number converted to numeric:")
print(df['doornumber'].value_counts().sort_index())
```

## 2.5 Handle Missing Values

```python
# Check for missing values
missing = df.isnull().sum()
missing = missing[missing > 0]
if len(missing) == 0:
    print("No missing values found.")
else:
    print("Missing values:")
    print(missing)
```

## 2.6 Handle Duplicates

```python
dup_count = df.duplicated().sum()
print(f"Duplicate rows: {dup_count}")
if dup_count > 0:
    df.drop_duplicates(inplace=True)
    print(f"Shape after removing duplicates: {df.shape}")
```

## 2.7 Check for Near-Zero Variance Columns

```python
# enginelocation is almost always 'front'
print("Engine Location distribution:")
print(df['enginelocation'].value_counts())
print(f"\n{df['enginelocation'].value_counts(normalize=True) * 100}")
```

```python
# Drop enginelocation — 98.5% is 'front', near-zero variance
df.drop(columns=['enginelocation'], inplace=True)
print(f"Dropped 'enginelocation'. Shape: {df.shape}")
```

## 2.8 Verify Cleaned Data

```python
print(f"Final shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nData types:")
print(df.dtypes)
print(f"\nMissing values: {df.isnull().sum().sum()}")
```
