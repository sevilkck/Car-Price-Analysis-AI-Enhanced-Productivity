## 1.3 Load the Dataset

```python
# Load dataset from Google Drive
file_path = '/content/drive/MyDrive/CarPrice_Assignment.csv'
df = pd.read_csv(file_path)
print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

## 1.7 Missing Values Check

```python
# Check for missing values
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing %', ascending=False)

if len(missing_df) == 0:
    print("No missing values found in the dataset!")
else:
    print("Missing values:")
    display(missing_df)
```

## 1.8 Duplicate Check

```python
# Check for duplicate rows
dup_count = df.duplicated().sum()
print(f"Number of duplicate rows: {dup_count}")
```

## 1.9 Unique Values per Column

```python
# Unique values for each column
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
    if df[col].dtype == 'object' and df[col].nunique() <= 20:
        print(f"  → Values: {df[col].unique().tolist()}")
    print()
```

## 1.10 Target Variable Overview

```python
# Quick look at target variable: price
print("Target Variable: price")
print(f"  Mean:   ${df['price'].mean():,.2f}")
print(f"  Median: ${df['price'].median():,.2f}")
print(f"  Min:    ${df['price'].min():,.2f}")
print(f"  Max:    ${df['price'].max():,.2f}")
print(f"  Std:    ${df['price'].std():,.2f}")
```
